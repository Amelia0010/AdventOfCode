```python
grid = [[*line] for line in input.split("\n")]
# Expand it
x = 0
while x < len(grid[0]):
    for line in grid:
        if line[x] == ".":
            continue
        break
    else:
        for line in grid:
            line.insert(x, ".")
        x += 1
    x += 1

y = 0
while y  < len(grid):
    if not ("#") in grid[y]:
        grid.insert(y, grid[y].copy())
        y += 1
    y += 1
```

```python
def calculateDistance(firstGalaxy: tuple, secondGalaxy: tuple) -> int:
    return abs(firstGalaxy[0] - secondGalaxy[0]) + abs(firstGalaxy[1] - secondGalaxy[1])
```

```python
galaxies = []
total = 0
for y, line in enumerate(grid):
    for x, char in enumerate(grid[y]):
        if char == "#":
            galaxy = (y, x)
            for otherGalaxy in galaxies:
                total += calculateDistance(galaxy, otherGalaxy)
            galaxies.append((y, x))
total
```

```python
EXPANSION = 999_999
EMPTY = "." * EXPANSION

emptyY = dict()

grid = input.split("\n")
# Expand it
x = 0
def getLength():
    return len(grid[0])

while x < getLength():
    for line in grid:
        if line[x] == ".":
            continue
        break
    else:
        for y, line in enumerate(grid):
            grid[y] = line[:x] + EMPTY + line[x:]
        x += EXPANSION
    x += 1

y = 0
while y  < len(grid):
    if not ("#") in grid[y]:
        lin = grid[y]
        for i in range(EXPANSION):
            emptyY[str(y+i)] = True
            grid.insert(y + i, lin)
        y += EXPANSION
    y += 1
```
