```python
lines = input.split("\n")
lineLength = len(lines[0])
singleLineInput = "".join(lines)

leftEdge = 3
rightEdge = lineLength - 3
topEdge = 3 * lineLength
bottomEdge = len(singleLineInput) - topEdge

UP = -lineLength
RIGHT = 1
DOWN = lineLength
LEFT = -1

startChars = ["X", "S"]

def posInLine(i: int):
    return i % lineLength

def check(position: int, dir: int) -> bool:
    return singleLineInput[position + dir] == "M" and singleLineInput[position + dir + dir] == "A" and singleLineInput[position + dir + dir + dir] == "S"
```

```python
xmasCount = 0
for i, char in enumerate(singleLineInput):

    if char != "X":
        continue

    linePos = posInLine(i)
    canCheckRight = linePos < (rightEdge)
    canCheckLeft = linePos >= leftEdge
    canCheckDown = i < bottomEdge
    canCheckUp = i > topEdge

    if canCheckRight:
        if check(i, RIGHT):
            xmasCount += 1
        if canCheckUp and check(i, UP + RIGHT):
            xmasCount += 1
        if canCheckDown and check(i, DOWN + RIGHT):
            xmasCount += 1

    if canCheckLeft:
        if check(i, LEFT):
            xmasCount += 1
        if canCheckUp and check(i, UP + LEFT):
            xmasCount += 1
        if canCheckDown and check(i, DOWN + LEFT):
            xmasCount += 1

    if (canCheckDown and check(i, DOWN)):
        xmasCount += 1

    if (canCheckUp and check(i, UP)):
        xmasCount += 1

print(f"Found {xmasCount} instances of festive xmas spirit!")
```
