```python
inputGrid = [[y for y in x] for x in input.split("\n")]

HAS_BEEN = "X"
OBSTACLE = "#"

UP = "UP"
RIGHT = "RIGHT"
DOWN = "DOWN"
LEFT = "LEFT"

lineLength = len(inputGrid[0])
height = len(inputGrid)

def isAtEdge(aX: int, aY: int, aDir: str) -> bool:
    if (aDir == UP):
        return aY == 0
    elif aDir == RIGHT:
        return aX == lineLength-1
    elif aDir == DOWN:
        return aY == height-1
    elif aDir == LEFT:
        return aX == 0
    else:
        print("WHAT")

def nextDir(aDir: str) -> str:
    if aDir == UP:
        return RIGHT
    elif aDir == RIGHT:
        return DOWN
    elif aDir == DOWN:
        return LEFT
    elif aDir == LEFT:
        return UP
    else:
        print("HUH")

def isObstacle(aX: int, aY: int) -> bool:
    return inputGrid[aY][aX] == OBSTACLE

def isNewTile(aX: int, aY: int) -> bool:
    return inputGrid[aY][aX] != HAS_BEEN

def nextTile(aX: int, aY: int, aDir: str):
    if aDir == UP:
        return aX, aY - 1
    elif aDir == RIGHT:
        return aX + 1, aY
    elif aDir == DOWN:
        return aX, aY + 1
    elif aDir == LEFT:
        return aX - 1, aY
    else:
        print("HMMM")

startX, startY = 0, 0
myDir = UP
tiles = 1

for i, line in enumerate(inputGrid):
    try:
        myIndex = line.index("^")
        startX, startY = myIndex, i
        line[myIndex] = "X"
        break
    except ValueError:
        pass

x,y = startX, startY

while not isAtEdge(x, y, myDir):
    myNewX, myNewY = nextTile(x, y, myDir)
    if isObstacle(myNewX, myNewY):
        myDir = nextDir(myDir)
    else:
        x, y = myNewX, myNewY
        if isNewTile(x, y):
            inputGrid[y][x] = HAS_BEEN
            tiles += 1
print(f"The guard will walk over {tiles} tiles!")
```
