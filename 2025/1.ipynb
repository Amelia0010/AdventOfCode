```python
dial = 50
oneCount = 0
twoCount = 0
with open("1.txt") as instructions:
    for instruction in instructions:
        left = instruction[0] == "L"
        ticks = (-1 if left else 1) * int(instruction[1:])
        dial = (dial + ticks) % 100
        if dial == 0:
            oneCount += 1
print("Part one: ", oneCount)
print("Part two: ", twoCount)
```
