```python
# Why the heck did I not just use a regex lmao
# This is literally just whether or not the string matches ([0-9]+)\1{$count} where $count = splits - 1
def isValid(number, splits):
    numAsString = str(number)
    length = len(numAsString)
    if length % splits != 0:
        return True
    else:
        splitIndex = int(length / splits)
        for index in range(0, splits-1):
            start = index*splitIndex
            halfway = start+splitIndex
            end = halfway + splitIndex
            if numAsString[start:halfway] != numAsString[halfway:end]:
                return True
        return False


import re
# So this is actually like thrice as slow. 
# Really cool tho
def efficienterValid(number, splits):
    numAsString = str(number)
    sliceSize = int(len(numAsString) / splits)
    return not re.search(f"^([0-9]{{{sliceSize}}})\\1{{{splits-1}}}$", numAsString)

# But if I don't do it for a specific split, then regex can be smart enough to figure it out itself and I don't have to loop through things
# This is actually more efficient, 
def efficienterValidForPartTwo(number):
    return not re.search("^([0-9]+)\\1{2,}$", str(number))

# And the same holds for part one.
# It is slightly annoying that I can't use the same function for both, but such is life. 
def efficienterValidForPartOne(number):
    return not re.search("^([0-9]+)\\1$", str(number))
```

```python
invalidsOne = 0
invalidsTwo = 0
with open("2.txt") as input:
    for idRange in input.readline().split(","):
        indices = idRange.split("-")
        for num in range(int(indices[0]), int(indices[1])+1):
            if not efficienterValidForPartOne(num):
                invalidsOne += num
                continue
            if not efficienterValidForPartTwo(num):
                invalidsTwo += num
print("Part one: ", invalidsOne)
print("Part two: ", invalidsOne + invalidsTwo)
```
