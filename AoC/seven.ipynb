```python
hands = [{"hand": x.split(" ")[0], "bid": x.split(" ")[1]} for x in input.split("\n")]
```

```python
types = [
    [r'([A-Z0-9])\1\1\1\1'],
    [r'([A-Z0-9])\1\1\1'],
    [r'([A-Z0-9])\1\1', r'([A-Z0-9])\1'],
    [r'([A-Z0-9])\1\1'],
    [r'([A-Z0-9])\1', r'([A-Z0-9])\1'],
    [r'([A-Z0-9])\1'],
    [r'[A-Z0-9]']
]

strengths = ["A", "K", "Q", "J", "T", "9", "8", "7", "6", "5", "4", "3", "2"]
```

```python
def getType(hand: str) -> int:
    sortedHand = "".join(sorted(hand))
    for index, typeArray in enumerate(types):
        handToSearch = sortedHand
        for type in typeArray:
            newHand = re.sub(type, "", handToSearch, 1)
            if newHand != handToSearch:
                handToSearch = newHand
                continue
            else:
                break
        else:
            return index
    else:
        return len(types)


for handDict in hands:
    handDict['type'] = getType(handDict['hand'])
    handDict['cards'] = [strengths.index(x) for x in handDict['hand']]
sortedHands = sorted(hands, key=lambda handDict: [handDict['type']] + handDict['cards'], reverse=True)
```

```python
total = 0
for i, handDict in enumerate(sortedHands):
    total += int(handDict['bid']) * (i + 1)
total
```
