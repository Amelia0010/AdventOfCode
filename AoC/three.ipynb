```python
grid = input.split("\n")
height = len(grid)
length = len(grid[0])
```

```python
def isSymbol(y: int, x: int) -> bool:
    return grid[y][x] != "." and not grid[y][x].isnumeric()

def nextToSymbol(y: int, x: int) -> bool:
    checkLeft = x > 0
    checkRight = x < length - 1
    checkUp = y > 0
    checkDown = y < height - 1
    nextToASymbol = False

    if checkUp:
        if checkLeft:
            nextToASymbol |= isSymbol(y-1, x-1)
        nextToASymbol |= isSymbol(y-1, x)
        if checkRight:
            nextToASymbol |= isSymbol(y-1, x+1)

    if checkLeft:
        nextToASymbol |= isSymbol(y, x-1)
    if checkRight:
        nextToASymbol |= isSymbol(y, x+1)

    if checkDown:
        if checkLeft:
            nextToASymbol |= isSymbol(y+1, x-1)
        nextToASymbol |= isSymbol(y+1, x)
        if checkRight:
            nextToASymbol |= isSymbol(y +1, x+ 1)

    return nextToASymbol
```

```python
x, y = 0, 0
sum = 0
while y < height:
    while x < length:
        numb = ""
        char = grid[y][x]
        if not char.isnumeric():
            x += 1
            continue
        else:
            hasSymbol = nextToSymbol(y, x)
            numb += char
            x += 1
            while x < length:
                char = grid[y][x]
                if not char.isnumeric():
                    x += 1
                    break
                numb += char
                hasSymbol |= nextToSymbol(y, x)
                x += 1
            if hasSymbol:
                sum += int(numb)

    y += 1
    x = 0
    continue
sum
```

```python
def isnum(y: int, x: int) -> bool:
    return grid[y][x].isnumeric()

def nextToNum(y: int, x: int) -> bool:
    checkLeft = x > 0
    checkRight = x < length - 1
    checkUp = y > 0
    checkDown = y < height - 1
    adjacentNums = 0

    if checkUp:
        if isnum(y-1, x):
            adjacentNums += 1
        else:
            if checkLeft and isnum(y-1, x-1):
                adjacentNums += 1
            if checkRight and isnum(y-1, x+1):
                adjacentNums += 1

    if checkLeft and isnum(y, x-1):
        adjacentNums += 1
    if checkRight and isnum(y, x+1):
        adjacentNums += 1

    if checkDown:
        if isnum(y+1, x):
            adjacentNums += 1
        else:
            if checkLeft and isnum(y+1, x-1):
                adjacentNums += 1
            if checkRight and isnum(y+1, x+1):
                adjacentNums += 1

    return adjacentNums == 2

def getNum(y: int, x: int) -> int:
    while x > 0:
        if isnum(y, x-1):
            x -= 1
        else:
            break
    numb = ""
    while x < length:
        if isnum(y, x):
            numb += grid[y][x]
            x += 1
        else:
            break



    return int(numb)

def countNums(y: int, x: int) -> int:
    if not nextToNum(y, x):
        return 0

    checkLeft = x > 0
    checkRight = x < length - 1
    checkUp = y > 0
    checkDown = y < height - 1
    adjacentNums = 1
    print(y, x)

    if checkUp:
        if isnum(y-1, x):
            adjacentNums *= getNum(y-1, x)
        else:
            if checkLeft and isnum(y-1, x-1):
                adjacentNums *= getNum(y-1, x-1)
            if checkRight and isnum(y-1, x+1):
                adjacentNums *= getNum(y-1, x+1)

    if checkLeft and isnum(y, x-1):
        adjacentNums *= getNum(y, x-1)
    if checkRight and isnum(y, x+1):
        adjacentNums *= getNum(y, x+1)

    if checkDown:
        if isnum(y+1, x):
            adjacentNums *= getNum(y+1, x)
        else:
            if checkLeft and isnum(y+1, x-1):
                adjacentNums *= getNum(y+1, x-1)
            if checkRight and isnum(y+1, x+1):
                adjacentNums *= getNum(y+1, x+1)

    return adjacentNums
```
